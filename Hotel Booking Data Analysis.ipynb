## Data Analysis and Visualizations 

```python
# calculating cancelled percentage

#cancelled_perc = df['is_canceled'].value_counts() this will count

# to return percentage use normalize
cancelled_perc = df['is_canceled'].value_counts(normalize = True) 
print(cancelled_perc)

#Visualize
plt.figure(figsize = (4,3.5))
plt.title("Reservation status count")
plt.bar(['Not canceled','Canceled'],df['is_canceled'].value_counts(), edgecolor = 'k', width=0.7)
plt.show()
```

```python
# Compare which hotels has a higher rate of cancellations
#Using seaborn
plt.figure(figsize = (8,4))
plt.title("Reservation Status in different hotels", size = 20)
plt.xlabel('Hotel')
plt.ylabel('number of reservations')
ax1 = sns.countplot(x='hotel', hue = 'is_canceled', data = df, palette = 'Blues')
plt.legend(['not canceled', 'canceled'])
plt.show()
```

```python
#Percent cancellation for Resort Hotel
resort_hotel = df[df['hotel']== "Resort Hotel"]
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
#Percent cancellation for City Hotel
city_hotel = df[df['hotel']== "City Hotel"]
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
#We have created a hypothesis that resort hotel will have more cancellation rate on the day when prices are higher

resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (20,8))
plt.title("Average Daily Rate in City and Resort Hotel", fontsize = 28)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel') #resort_hotel.index is the day
plt.plot(city_hotel.index, city_hotel['adr'], label = "City Hotel")
plt.legend(fontsize = 20)
plt.show()
```

```python
#PLot a graph for months to find which months have the highest cancellation

#Note we do not have a specific month column hence we extract months:
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
plt.title("Reservation Status per Month", size = 20)
plt.xlabel("month")
plt.ylabel("number of reservations")
ax1 = sns.countplot(x= 'month', hue = 'is_canceled', data = df, palette = 'bright')
plt.show()
```

```python
#Now we will find whether there is any relationship between price and cancellation rate plot a graph
plt.title('ADR per month', fontsize = 30)

#we are using .reset_index() to take out the month column
sns.barplot('month', 'adr', data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.show()

#The graph confirms that our hypothesis is correct. When ADR is high the cancellation too is high 
```

```python
#Compare the countries which have higher cancellation rate(We are taking top 10)
cancelled_data = df[df['is_canceled']== 1]
top_10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize= (8,8))
plt.title("Top 10 countries with reservation canceled ")
plt.pie(top_10_country, autopct = '%.2f', labels = top_10_country.index)
plt.show()
```

```python
#Let us check where are the customers coming from
df['market_segment'].value_counts()
```

```python
#We will find reservation cancelled Percentage through which mode
cancelled_data['market_segment'].value_counts(normalize = True)
```

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace = True)
cancelled_df_adr.sort_values("reservation_status_date", inplace= True)

not_cancelled_df = df[df['is_canceled']== 0]
not_cancelled_df_adr = not_cancelled_df.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace = True)
not_cancelled_df_adr.sort_values("reservation_status_date", inplace= True)

plt.figure(figsize = (20,6))
plt.title("Average Daily Rate")
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'], label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'], label = 'cancelled')
plt.legend()
plt.show()
```

```python
#We can see there is inconsistency in the data
#We will take data from 2016 to 2017 since it is the most consistent

cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016') & (cancelled_df_adr['reservation_status_date']<'2017-09')] 
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016') & (not_cancelled_df_adr['reservation_status_date']<'2017-09')]
```

```python
#Plotting the graph again
plt.figure(figsize = (20,6))
plt.title("Average Daily Rate", fontsize = 30)
plt.xlabel("Month",fontsize = 18)
plt.ylabel("Average Daily Rate", fontsize = 18)
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'], label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'], label = 'cancelled')
plt.legend(fontsize = 20)
plt.show()
```
